# ANONYMISATION DES DONNEES

```python
df_anonymise = df.copy()
```

```python
df_anonymise["phone"] = df_anonymise["phone"].apply(mask_data)

df_anonymise["emp_email"] = df_anonymise["emp_email"].apply(pseudo_data)

df_anonymise["texte"] = df_anonymise["texte"].apply(delete_data)

df_anonymise["texte1"] = df_anonymise["texte1"].apply(hash_data)

df_anonymise["mot"] = df_anonymise["mot"].apply(substitute_data, substitution_map=substitution_map)

df_anonymise["date"] = df_anonymise["date"].apply(generalize_data, level= 10)

df_anonymise["mot1"] = df_anonymise["mot1"].apply(randomize_data)

df_anonymise["mot2"] = df_anonymise["mot2"].apply(encrypt_data)
```
